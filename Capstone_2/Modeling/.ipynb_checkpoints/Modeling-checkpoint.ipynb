## Importing relevant libararies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Machine Learning tools
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import statsmodels.api as sm

# Set seaborn style
sns.set_style('whitegrid')

# Suppress future warnings
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
warnings.simplefilter(action='ignore', category=UserWarning)
```

# Load the data

```python
explored_data = "Explored_data_copy"
df = pd.read_csv(explored_data)
df.head()
```

# Preprocessing

```python
# Print the data type of each column
df.dtypes
```

### Create dummy or indicator features for categorical variables

```python
# Get only object columns
df_cats = df.select_dtypes('object')

# Seperate features from target feature
X = df.drop(columns=['price'], axis=1)
y = df[['price']]

# Get n-1 dummies for all categorical variables
X = pd.get_dummies(X, columns=df_cats.columns, drop_first=True, prefix='C')
X.head()
```

# Fit models with a training dataset

### Split data into testing and training subsamples

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=1)
```

```python
# We now want to check the shape of the X train, y_train, X_test and y_test to make sure the proportions are right. 
print(X_train.shape, X_test.shape)
print(y_train.shape, y_test.shape)
```

### Standardize the magnitude of numeric features using a scaler

```python
# View range of data
X_train.describe()
```

```python
# Initialize scaler
scaler = StandardScaler()

# Build scaler based on training data
X_train_scaled = scaler.fit_transform(X_train)

# Apply (but not fit) scaler to test data
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
X_test_scaled
```

## Making a Linear Regression model: First model
Sklearn has a [LinearRegression()](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html) function built into the linear_model module. 

```python
from sklearn import linear_model, preprocessing # We'll be using that to make our regression model. 
```

```python
# Create the model: make a variable using linear_model.LinearRegression
# The normalize parameter would be set to True if our variables were not already scaled, from previous steps.
rModel = linear_model.LinearRegression(normalize=False)
```

```python
# We now want to train the model on our test data
rModel.fit(X_train_scaled, y_train)
```

```python
# Evaluate the model
rModel.score(X_train_scaled, y_train)
```

The above score is called R-Squared coefficient, or the "coefficient of determination". It's basically a measure of how successfully our model predicts the variations in the data away from the mean: 1 would mean a perfect model that explains 100% of the variation. At the moment, our model explains about 92% of the variation from the mean. This is suspiciously high, maybe there is more work to do!

```python
# Use the model to make predictions about our test data
y_predict = rModel.predict(X_test)
```

```python
# Let's plot the predictions against the actual results
plt.figure(figsize=(15,7.5))
plt.scatter(y_test, y_predict)
_ = plt.xlabel('Tested price')
_ = plt.ylabel('Predicted price')
_ = plt.title('Test vs Prediction')
```

The above scatterplot represents how well the predictions match the actual results. 

Along the x-axis, we have the actual price, and along the y-axis we have the predicted value for the price.

There is a visible positive correlation, as the model has not been totally unsuccesful, but it's clear that it is not maximally accurate: diamonds with an actual price of just over 11250 have been predicted as having prices from about -4000 to about 6,000.

Let's build a similar model using a different package, to see if we get a better result that way.

## Making a Linear Regression model: Second model: Ordinary Least Squares (OLS)

```python
# Use statsmodel
import statsmodels.api as sm

# Create the test and train sets. Here, we can do things slightly different using other code.
X = sm.add_constant(X)
```

```python
# Do the same as before for y
y = df[['price']]

# Split the data using train_test_split()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.25, random_state=123)
```

```python
# Create the model
rModel2 = sm.OLS(y_train, X_train)

# Fit the model
rModel2_results = rModel2.fit()
```

```python
# Evaluate the model
rModel2_results.summary()
```

One of the great things about Statsmodels (sm) is the large amount of information from the summary() method. 

There are lots of values, but one of the most important: the R-squared score is 0.919, the same as what it was with the previous model. It's the same value as the score from sklearn, because they've both used the same algorithm on the same data.

Here's a useful link for reference: https://www.theanalysisfactor.com/assessing-the-fit-of-regression-models/

```python
# Now it's time to make predictions using the new model
y_pred = rModel2_results.predict(X_test)
```

```python
# Plot predictions using scatter plot again
plt.figure(figsize=(15,7.5))
plt.scatter(y_test, y_pred)

# This time add line for perfect correlation
_ = plt.plot([x for x in range(0,17500)], [x for x in range(0,17500)], color='red')

_ = plt.title("Model 2 predictions vs actual values")
_ = plt.xlabel("Actual")
_ = plt.ylabel("Predicted")
```

We can see that we have a great match between our data and our predictions, and we can see that the shape of the data points is very similar to the red line. 

We can check another metric as well - the RMSE (Root Mean Squared Error). The MSE is defined by Professor Spiegelhalter on p.393 of *AoS*, and the RMSE is just the square root of that value. This is a measure of the accuracy of a regression model. Very simply put, it's formed by finding the average difference between predictions and actual values.

```python
# Define a function to check the RMSE. 
def rmse(predictions, targets):
    return np.sqrt(((predictions - targets)** 2).mean())
```

```python
# Get predictions from rModel2
y_pred = rModel2_results.predict(X_test)

# Put the predictions & actual values into a dataframe
matches = pd.DataFrame(y_test)
matches.rename(columns={'price':'actual'}, inplace=True)
matches['predicted'] = y_pred

matches
```

```python
rmse(matches['actual'], matches['predicted'])
```

The RMSE tells us how far, on average, our predictions were mistaken. An RMSE of 0 would mean we were making perfect predictions. 0.5 signifies that we are, on average, about 0.6 of a unit of price away from the correct answer.
